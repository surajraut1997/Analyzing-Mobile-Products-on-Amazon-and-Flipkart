# Data Scraping - Amazon Mobile Cateory 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Product_name = []
Price = []
Review = []

# Loop through multiple pages
for i in range(1, 20):
    url = f"https://www.amazon.in/s?k=mobiles+under+15000&crid=32CMZLODWXBBN&page={i}&qid=1683868917&sprefix=%2Caps%2C1130&ref=sr_pg_{i}"
    print(url)

    # Set the User-Agent header to mimic a web browser
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"
    }

    # Send a GET request to the URL
    r = requests.get(url, headers=headers)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(r.text, "html.parser")

    # Find the container with the search results
    box = soup.find("div", class_="s-result-list")

    # Find all individual items in the container
    items = box.find_all("div", class_="s-result-item")

    # Extract relevant information for each item
    for item in items:
        # Extract the product name
        name = item.find("h2", class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")
        if name is not None:
            name = name.text.strip()

        # Extract the price
        price = item.find("span", class_="a-price-whole")
        if price is not None:
            price = price.text

        # Extract the review rating
        rating = item.find("span", class_="a-icon-alt")
        if rating is not None:
            rating = rating.text.split()[0]

        # Add the extracted information to respective lists
        if name and price and rating:
            Product_name.append(name)
            Price.append(price)
            Review.append(rating)

# Print the lengths of the lists to verify the data
print(len(Product_name))
print(len(Price))
print(len(Review))

# Create a DataFrame with the extracted data
df = pd.DataFrame({
    "Product Name": Product_name,
    "Price": Price,
    "Review": Review
})

# Check for missing values and replace them with 0
df = df.fillna(0)

# Export the DataFrame as a CSV file
df.to_csv("Amazon Reviews20.csv", index=False)

# Display the first few rows of the DataFrame
print(df.head())
```
